```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
path = r"C:\Users\Mohd Junaid Mahmood\Desktop/TSLA.csv"
df = pd.read_csv(path)
df.head()
```

```python
df.columns
```

```python
plt.plot(df['Date'], df['Open'], label='Open')
plt.plot(df['Date'], df['High'], label='High')
plt.plot(df['Date'], df['Low'], label='Low')
plt.xlabel('Date')
plt.ylabel('Volume')
plt.title('Popularity of Volume terms by date')
plt.grid(True)
plt.show()
```

```python
fig, axes = plt.subplots(2,2)
axes[0, 0].hist(df['Volume'])
axes[0, 1].scatter(df['Date'], df['Adj Close'])
axes[1, 0].plot(df['Open'], df['Close'])
axes[1, 1].plot(df['High'], df['Low'])
```
